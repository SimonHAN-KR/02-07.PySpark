#### Why does Spark exist? 
- Spark exists to clean | transform | analyze | make predictions on _big data_. 

- Especially suited to things like Natural Language Processing and Image Processing (Computer Vision). 

- Spark splits your big data into smaller partitions, and stores it 'in-memory' across many physical or virtual servers. 

- Spark performs computation on each partition in parallel, making it a lot faster. 



---
#### Why have Microsoft built their end-to-end data analytics platform around it? 
- The data world have converged on Spark. It is used by thousands of companies, including 80% of the Fortune 500. 
- Spark is relatively simple to learn and modular.
